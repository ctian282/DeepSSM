```python
import jax
import jax.numpy as jnp
import numpy as np
import os
import matplotlib.pyplot as plt

# uncomment this can add your own when library such as libdevice.10.bc cannot be found
#os.environ["XLA_FLAGS"] = "--xla_gpu_cuda_data_dir="+os.environ['CONDA_DIR']

from deepssm import spectrum
```

```python
# uncomment this if only cpu is going to be used
# (use GPU by default if available)
jax.config.update('jax_platform_name', 'cpu')
sp = spectrum.GWSpectrum()
```

```python
%%time
# setting phase transition parameters
vw, alpha, Htau, HR = 0.1, 0.3, 0.008, 0.003
Ts = 300 # T_*
f = np.logspace(-4, -1, 100) # frequencies
GWs= sp.get_sepctrum(f, vw, alpha, Htau, HR, Ts, g_st=100, extrapolate=False)
```

```python
plt.loglog(f, GWs)
plt.xlabel('f(Hz)')
plt.ylabel('Omega_GW')
plt.show()
```
